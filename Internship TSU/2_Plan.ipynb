               #          **План исследования для поставленной задачи:**

**Заказчик -**  
IT-компания, последние 15 лет успешно занимавшаяся разработкой и консалтингом на российском рынке.

**Методология исследования, этапы исследования -** 

Методология CRISP-DM:



**1)**	Анализ данных                                                                                                   
•	Сбор данных – анализ всех источников, если у заказчика их недостаточно, организовать сбор данных с агрегаторов онлайн курсов и популярных баз вакансий                                                                                              
•	Описание данных – описать кол-во строк, и вес файла собранных данных                                                             
•	Изучение данных на предмет соответствия их цели анализа, ищем ответ на вопрос как эти данные помогут решить задачу.             
•	Проверка качества данных на отсутствие пропущенной информации, опечатки.
                                                                                                    
**2)**	Подготовка данных                                                                                                           
•	Выборка данных собранных с агрегаторов онлайн курсов, популярных баз вакансий, не релевантные данные, которые не помогают решить задачу удаляются                                                                                                           
•	Очистка данных                                                                                                  
    - заполняем пропущенные значения или удаляемих                                                                         
    - исправляем ошибки в данных или если они не нанесут вред результату, удаляем их                                            
    - исправляем кодировку если она нарушена                                                                            
•	Генерация данных – проводим агрегацию максимальной и минимальной оценки онлайн курса конкурента, агрегацию средней зарплаты в сферах IT разработки и консалтинга.                                                                           
•	Интеграция данных а именно создание таблиц (данные с агрегаторов онлайн курсов и данные с базы вакансий) для работы с ними.

**3)**	Моделирование                                                                                               
•	Выбор алгоритмов:
- Достаточно ли данных, поскольку сложные модели как правило требуют большей выборки?                                               
- Сможет ли модель обработать пропуски данных (какие-то реализации алгоритмов умеют работать с пропусками, 
какие-то нет)?                                                                                                          
- Сможет ли модель работать с имеющимися типами данных или необходима конвертация?                                                   
•	Планирование тестирования                                                                                       
•	Обучение моделей                                                                                            

**4)**	Оценка результатов                                                                      
•	Оценка результатов – На сколько хорошо полученные результаты решают бизнес – задачу?                                
•	Оценка процесса Выявляем сильные и слабые стороны которые были в проекте.                                       
•	Определение следующих шагов Дальше улучшать проект или результат удовлетворит заказчика?                                    

**5)**	Внедрение 
•	Подготовка отчета, выводов и рекомендаций                                                           
•	Ревью проекта

Занять свою нишу в  онлайн-образовании, благодаря качественным онлайн-курсам  для подготовки IT специалистов с знаниями востребованными на рынке труда.
