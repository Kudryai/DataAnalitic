**Создание облака слов из чата телеграм.**
В данном примере был импортирован чат группы обучения Дата аналитик Томского государственного университета - result.json
Далее с помощью скрипта Python (CloudTelegram.py) он был очищен и экспортирован в формате txt.
Следущий шаг с помощью билбиотек matplotlib & wordcloud в файле fun.ipynb, он был преобразован в облако слов.

```python
import matplotlib.pyplot as plt
%matplotlib inline

from wordcloud import WordCloud, STOPWORDS

tgu = open('/mnt/c/Users/user/Desktop/DataAnalyst/DataAnalyst/DataAnalyst/Create_wordcloud_Telegram/out.txt', 'r', encoding = 'utf-8')
tgu1 = tgu.read()
tgu.close()

wordCloud = WordCloud(width = 10000, height = 10000, random_state=1, background_color='black', colormap='Set2', collocations=False).generate(tgu1)

plt.figure(figsize=(5,5))
plt.imshow(wordCloud)
```
